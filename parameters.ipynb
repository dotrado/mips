```python
import numpy as np
#from sympy import *
from sympy import integrate, log, exp, oo
from sympy.abc import x
from sklearn.grid_search import ParameterGrid
r = Symbol('r')
d = Symbol('d')

def phi(y):
    return integrate(exp(-x**2 /2)/sqrt(2*pi), (x, -oo, y))

def F_r(r, d):
    return (1 - 2*phi(-r/d) - 2*(1 - exp(-(r/d)**2 /2))/sqrt(2*pi)*d/r)

def rho(m, S_0, U, c, R):
    return (log(F_r(R, sqrt(1 + m/4 - 2*S_0 + U**(2**m + 1))))/ log(F_r(R, sqrt(1 + m/4 - 2*c*S_0))))



U = np.arange(0.75, 0.95, 0.001).tolist()
m = [2, 3, 4, 5]
r = np.arange(1.5, 3, 0.01).tolist()
c = np.arange(0.0, 1.0, 0.001).tolist()
param_grid = {'param1':U, 'param2':r, 'param3':m}
grid = ParameterGrid(param_grid)
S = [0.5, 0.6, 0.7, 0.8, 0.9]
val_S = []
val_1 = []

for p_ in c:
    val_1 = []
    for s0 in S:
        fval = 100
        for m0 in m:
            for U0 in U:
                for r0 in r:
                    print("Hi")
                    fval1 = rho(m0, s0, U0, p_, r0)
                    if(fval1 < fval):
                        fval = fval1
        val_1.append(fval)
    val_S.append(val_1)
```
